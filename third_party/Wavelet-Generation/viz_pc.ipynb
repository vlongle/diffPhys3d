```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import trimesh

def load_and_visualize_obj(obj_path):
    """
    Load an .obj file and visualize it as a point cloud
    
    Args:
        obj_path (str): Path to the .obj file
    """
    # Load the mesh using trimesh
    mesh = trimesh.load(obj_path)
    
    # Extract vertices (points) from the mesh
    points = mesh.vertices
    
    # Create a 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot the points
    ax.scatter(
        points[:, 0],  # x coordinates
        points[:, 1],  # y coordinates
        points[:, 2],  # z coordinates
        c='blue',
        marker='.',
        s=1,
        alpha=0.5
    )
    
    # Set labels and title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Point Cloud Visualization: {obj_path}')
    
    # Auto-scale the axes
    max_range = np.max([
        np.ptp(points[:, 0]),
        np.ptp(points[:, 1]),
        np.ptp(points[:, 2])
    ])
    mid_x = np.mean([np.min(points[:, 0]), np.max(points[:, 0])])
    mid_y = np.mean([np.min(points[:, 1]), np.max(points[:, 1])])
    mid_z = np.mean([np.min(points[:, 2]), np.max(points[:, 2])])
    ax.set_xlim(mid_x - max_range/2, mid_x + max_range/2)
    ax.set_ylim(mid_y - max_range/2, mid_y + max_range/2)
    ax.set_zlim(mid_z - max_range/2, mid_z + max_range/2)
    
    plt.tight_layout()
    plt.show()
    
    return mesh
```

```python
obj_path = "debug/2025-04-04_11-36-24_Network-Marching-Cubes-Diffusion-Gen/200_265.obj"
load_and_visualize_obj(obj_path)
```
